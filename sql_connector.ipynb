## MySQL Database Connector

Connects to a local MySQL database and loads the banking clients dataset into a Pandas DataFrame for downstream analysis.

**Prerequisites:**
- MySQL server running on `localhost:3306`
- A `.env` file in the project root containing `DB_HOST`, `DB_PORT`, `DB_USER`, and `DB_PASSWORD`
- The `banking` schema with the `banking-clients` table imported from `banking-clients.csv`

```python
import os
import mysql.connector
import pandas as pd
from dotenv import load_dotenv

load_dotenv()  # reads credentials from .env

cnx = mysql.connector.connect(
    host=os.environ.get("DB_HOST", "127.0.0.1"),
    port=int(os.environ.get("DB_PORT", 3306)),
    user=os.environ.get("DB_USER", "root"),
    password=os.environ.get("DB_PASSWORD"),
)
```

```python
query = "SELECT * FROM banking.`banking-clients`"
```

```python
df = pd.read_sql(query, cnx)
```

```python
cnx.close()
```
